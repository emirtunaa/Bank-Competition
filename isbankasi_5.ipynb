# Converting Target Column to Proper Format

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np

data = pd.read_parquet('/Users/emirtuna/Desktop/train_final.parquet', engine='pyarrow')

# Ensure the 'target' column contains lists of floats
data['target'] = data['target'].str.replace('menu', '').str.split(',').apply(lambda menu_list: np.array([float(menu) for menu in menu_list], dtype=object))

data[['target_1', 'target_2', 'target_3']] = pd.DataFrame(data['target'].tolist(), index=data.index)


max_menu_number = 9

# Function to convert a list of menu numbers to a binary string
def convert_to_binary_list(menu_list):
    # Create an empty binary string
    binary_string = ['0'] * max_menu_number

    # Set '1' in the binary string for recommended menus
    for menu_number in menu_list:
        binary_string[int(menu_number) - 1] = '1'

    # Join the binary string to form the final representation
    return ''.join(binary_string)

# Create the 'target_result' column by applying the function to the 'target' column
data['target_result'] = data['target'].apply(convert_to_binary_list)

# Drop the original 'target' column
data.drop(columns=['target'], inplace=True)

data.head()
#data.columns
```
